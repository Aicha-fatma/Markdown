```python
from tkinter import *
from tkinter.font import Font

from tkinter.ttk import Progressbar
from tkinter import ttk
from PIL import Image
#####################
from tkinter import *
from tkinter import filedialog
from tkinter import font
from tkinter import colorchooser
import os, sys
import win32print
import win32api

class MainApp():
    def __init__(self):
        super().__init__()
        self = Tk()
        self.title('Text Editor Markdown')
        self.geometry("1200x680")
        self.state('zoomed')
        self.configure(bg='thistle')

        global open_status_name
        open_status_name = False
        global selected
        selected = False

        #New file
        def new_file():
            text.delete("1.0", END)
            self.title('New File')
            status_bar.config(text="New Fill        ")

            global open_status_name
            open_status_name = False    
        #Open 
        def open_file():
            text.delete("1.0", END)
            text_file =filedialog.askopenfilename(initialdir="", title="Open File", filetypes="")
            if text_file:
                global open_status_name
                open_status_name = text_file
            name = text_file
            status_bar.config(text=f'{name}       ')
            name = name.replace("C:/Users/DELL/Desktop/", "")
            self.title(f'{name}')

            text_file = open(text_file, 'r')
            stuff = text_file.read()
            text.insert(END, stuff)
            text_file.close()


        #SAve as
        def save_as_file():
            text_file = filedialog.asksaveasfilename(defaultextension=" .*", initialdir="", title="Save File", filetypes="")
            if text_file:
                name = text_file
                status_bar.config(text=f'Saved: {name}       ')
                name = name.replace("C:/Users/DELL/Desktop/", "")
                root.title(f'{name}')
                text_file = open(text_file, 'w')
                text_file.write(text.get("1.0", END))
                text_file.close()
        #save         
        def save_file():
            global open_status_name
            if open_status_name:
                text_file = open(open_status_name, 'w')
                text_file.write(text.get("1.0", END))
                text_file.close()
                status_bar.config(text=f'Saved: {open_status_name}        ')
            else:
                save_as_file()
        #cut
        def cut_text(e):
            global selected
            if e:
                selected = self.clipboard_get()
            else:
                if text.selection_get():
                    selected = text.selection_get()
                    text.delete("sel.first", "sel.last")
                    self.clipboard_clear()
                    self.clipboard_append(selected)

        #copy
        def copy_text(e):
            global selected
            if e:
                selected = root.clipboard_get()
            if text.selection_get():
                selected = text.selection_get()
                self.clipboard_clear()
                self.clipboard_append(selected)




        #past
        def paste_text(e):
            global selected
            if e:
                selected = root.clipboard_get()
            else:
                if selected:
                    position = text.index(INSERT)
                    text.insert(position, selected)
        #Bold text
        def bold():
            bold_font = font.Font(text, text.cget("font"))
            bold_font.configure(weight="bold")
            text.tag_configure("bold", font=bold_font)
            current_tags = text.tag_names("sel.first")

            if "bold" in current_tags:
                text.tag_remove("bold", "sel.first", "sel.last")
            else:
                text.tag_add("bold", "sel.first", "sel.last")
        #bold_italic
        def bold_italic():
            bold_italic_font = font.Font(text, text.cget("font"))
            bold_italic_font.configure(weight="bold", slant="italic")
            text.tag_configure("bold_italic", font=bold_italic_font)
            current_tags = text.tag_names("sel.first")

            if "bold_italic" in current_tags:
                text.tag_remove("bold_italic", "sel.first", "sel.last")
            else:
                text.tag_add("bold_italic", "sel.first", "sel.last")
            
        #Italic
        def italic():
            italic_font = font.Font(text, text.cget("font"))
            italic_font.configure(slant="italic")
            text.tag_configure("italic", font=italic_font)
            current_tags = text.tag_names("sel.first")

            if "italic" in current_tags:
                text.tag_remove("italic", "sel.first", "sel.last")
            else:
                text.tag_add("italic", "sel.first", "sel.last")
        ###titre
        def titre():
            titre_font = font.Font(text, text.cget("font"))
            titre_font.configure(size=30, weight="bold")
            text.tag_configure("title", font=titre_font)
            current_tags = text.tag_names("sel.first")

            if "title" in current_tags:
                text.tag_remove("title", "sel.first", "sel.last")
            else:
                text.tag_add("title", "sel.first", "sel.last")

        #sous titre
        def sous_titre():
            sous_titre_font = font.Font(text, text.cget("font"))
            sous_titre_font.configure(size=20, weight="bold")
            text.tag_configure("sous_title", font=sous_titre_font)
            current_tags = text.tag_names("sel.first")

            if "sous_title" in current_tags:
                text.tag_remove("sous_title", "sel.first", "sel.last")
            else:
                text.tag_add("sous_title", "sel.first", "sel.last")

        #noemal
        def normal():
            normal_font = font.Font(text, text.cget("font"))
            normal_font.configure(weight="normal", slant="roman")
            text.tag_configure("nermal", font=normal_font)
            current_tags = text.tag_names("sel.first")

            if "normal" in current_tags:
                text.tag_remove("normal", "sel.first", "sel.last")
            else:
                text.tag_add("normal", "sel.first", "sel.last")
        #sousligne
        def sousligne():
            sousligne_font = font.Font(text, text.cget("font"))
            sousligne_font.configure(underline=1)
            text.tag_configure("sousligne", font=sousligne_font)
            current_tags = text.tag_names("sel.first")

            if "sousligne" in current_tags:
                text.tag_remove("sousligne", "sel.first", "sel.last")
            else:
                text.tag_add("sousligne", "sel.first", "sel.last")
        #strike
        def strike():
            strike_font = font.Font(text, text.cget("font"))
            strike_font.configure(overstrike=1)
            text.tag_configure("strike", font=strike_font)
            current_tags = text.tag_names("sel.first")

            if "strike" in current_tags:
                text.tag_remove("strike", "sel.first", "sel.last")
            else:
                text.tag_add("strike", "sel.first", "sel.last")
        
        #bold_sligne
        def b_s():
            bold_font = font.Font(text, text.cget("font"))
            bold_font.configure(weight="bold", underline=1)
            text.tag_configure("bold_sousligne", font=bold_font)
            current_tags = text.tag_names("sel.first")
            if "bold_sousligne" in current_tags:
                text.tag_remove("bold_sousligne", "sel.first", "sel.last")
            else:
                text.tag_add("bold_sousligne", "sel.first", "sel.last")
            

        #bole_strink
        def b_st():
            bold_font = font.Font(text, text.cget("font"))
            bold_font.configure(weight="bold", overstrike=1)
            text.tag_configure("bold_strike", font=bold_font)
            current_tags = text.tag_names("sel.first")
            if "bold_strike" in current_tags:
                text.tag_remove("bold_strike", "sel.first", "sel.last")
            else:
                text.tag_add("bold_strike", "sel.first", "sel.last")
            
        #itali_souligne
        def it_s():
            bold_font = font.Font(text, text.cget("font"))
            bold_font.configure(slant="italic", underline=1)
            text.tag_configure("italic_sousligne", font=bold_font)
            current_tags = text.tag_names("sel.first")
            if "italic_sousligne" in current_tags:
                text.tag_remove("italic_sousligne", "sel.first", "sel.last")
            else:
                text.tag_add("italic_sousligne", "sel.first", "sel.last")
        #itali_skt
        def it_sk():
            bold_font = font.Font(text, text.cget("font"))
            bold_font.configure(slant="italic", overstrike=1)
            text.tag_configure("italic_strike", font=bold_font)
            current_tags = text.tag_names("sel.first")
            if "italic_strike" in current_tags:
                text.tag_remove("italic_strike", "sel.first", "sel.last")
            else:
                text.tag_add("italic_strike", "sel.first", "sel.last")  
                
        #b_it_s        
        def b_it_s():
            bold_italic_font = font.Font(text, text.cget("font"))
            bold_italic_font.configure(weight="bold", slant="italic", underline=1)
            text.tag_configure("bold_italic_souligne", font=bold_italic_font)
            current_tags = text.tag_names("sel.first")

            if "bold_italic_souligne" in current_tags:
                text.tag_remove("bold_italic_souligne", "sel.first", "sel.last")
            else:
                text.tag_add("bold_italic_souligne", "sel.first", "sel.last")   
                
        #b_it_sk
        def b_it_sk():
            bold_italic_font = font.Font(text, text.cget("font"))
            bold_italic_font.configure(weight="bold", slant="italic", overstrike=1)
            text.tag_configure("bold_italic_strik", font=bold_italic_font)
            current_tags = text.tag_names("sel.first")

            if "bold_italic_strik" in current_tags:
                text.tag_remove("bold_italic_strik", "sel.first", "sel.last")
            else:
                text.tag_add("bold_italic_strik", "sel.first", "sel.last")
        #coleur
        def size():
            s=[8,10,12,14,16,18,20,24,36,40,48]
            for size in s:
                size_f('end', size)

        
        #coleur
        def text_color():
            my_color = colorchooser.askcolor()[1]
            if my_color:

                color_font = font.Font(text, text.cget("font"))
                text.tag_configure("colored", font=color_font, foreground=my_color)
                current_tags = text.tag_names("sel.first")

                if "colored" in current_tags:
                    text.tag_remove("colored", "sel.first", "sel.last")
                else:
                    text.tag_add("colored", "sel.first", "sel.last")


        #color background
        def bg_color():
            my_color = colorchooser.askcolor()[1]
            if my_color:
                text.config(bg=my_color)
        #all text
        def all_text_color():
            my_color = colorchooser.askcolor()[1]
            if my_color:
                text.config(fg=my_color)

        #print file
        def print_file():
            #printer_name = win32print.GetDefaultPrinter()
            #status_bar.config(text=printer_name)
            file_to_print = filedialog.askopenfilename(initialdir="", title="Open File", filetypes="")
            if file_to_print:
                win32api.ShellExecute(0, "print", file_to_print, None, ".", 0)

        #night mode on
        def night_on():
            main_color = "#000000"
            text_color = "green"
            sc = "#373737"
            self.config(bg=main_color)
            status_bar.config(bg=main_color, fg=text_color)
            text.config(bg=sc)
            toolbar_frame.config(bg=main_color)
            bold_button.config(bg=sc, fg=text_color)
            italic_button.config(bg=sc, fg=text_color)
            redo_button.config(bg=sc, fg=text_color)
            undo_button.config(bg=sc, fg=text_color)
            b_s.config(bg=sc, fg=text_color)
            b_st.config(bg=sc, fg=text_color)
            normal_button.config(bg=sc, fg=text_color)
            bold_italic_button.config(bg=sc, fg=text_color)
            strike_button.config(bg=sc, fg=text_color)
            sous_titre_button.config(bg=sc, fg=text_color)
            titre_button.config(bg=sc, fg=text_color)
            sousligne_button.config(bg=sc, fg=text_color)
            color_text_button.config(bg=sc, fg=text_color)
            file_menu.config(bg=main_color, fg=text_color)
            edit_menu.config(bg=main_color, fg=text_color)
            color_menu.config(bg=main_color, fg=text_color)
            options_menu.config(bg=main_color, fg=text_color)
            it_s.config(bg=sc, fg=text_color)
            it_sk.config(bg=sc, fg=text_color)
            b_it_s.config(bg=sc, fg=text_color)
            b_it_sk.config(bg=sc, fg=text_color)
            size_f.config(bg=sc, fg=text_color)


        #nigth mode off
        def night_off():
            main_color = "SystemButtonFace"
            text_color = "SystemButtonFace"
            sc = "black"
            self.config(bg=main_color)
            status_bar.config(bg=main_color, fg=text_color)
            text.config(bg="white")
            toolbar_frame.config(bg=main_color)
            bold_button.config(bg="white")
            italic_button.config(bg="white")
            redo_button.config(bg="white")
            undo_button.config(bg="white")
            
            normal_button.config(bg="white")
            bold_italic_button.config(bg="white")
            strike_button.config(bg="white")
            sous_titre_button.config(bg="white")
            titre_button.config(bg="white")
            sousligne_button.config(bg="white")
            color_text_button.config(bg="white")
            file_menu.config(bg=main_color)
            edit_menu.config(bg=main_color)
            color_menu.config(bg=main_color)
            options_menu.config(bg=main_color)




        #toolbar
        toolbar_frame =Frame(self)
        toolbar_frame.pack(fill=X)

        #Frame
        frame = Frame(self)
        frame.pack(pady=5)



        #scroll

        text_scroll =Scrollbar(frame)
        text_scroll.pack(side=RIGHT, fill=Y)

        #horizantal
        hor_scroll = Scrollbar(frame, orient='horizontal')
        hor_scroll.pack(side=BOTTOM, fill=X)

        #text
        text = Text(frame, width=97, height=25, font=("Helvetica", 16), selectbackground="blue", selectforeground="black", undo=True, yscrollcommand=text_scroll.set, wrap="none", xscrollcommand=hor_scroll.set)
        text.pack()

        #configration scroll
        text_scroll.config(command=text.yview)
        hor_scroll.config(command=text.xview)


        #menu
        my_menu = Menu(self)
        self.config(menu=my_menu)
        #fill menu

        file_menu = Menu(my_menu, tearoff=False)
        my_menu.add_cascade(label="File", menu=file_menu)
        file_menu.add_command(label="New", command=new_file)
        file_menu.add_command(label="Open", command=open_file)
        file_menu.add_command(label="Save", command=save_file)
        file_menu.add_command(label="Save As", command=save_as_file)
        file_menu.add_separator()
        file_menu.add_command(label="Print File", command=print_file)
        file_menu.add_separator()
        file_menu.add_command(label="Exit", command=self.quit)

        #Etid menu
        edit_menu = Menu(my_menu, tearoff=False)
        my_menu.add_cascade(label="Edit", menu=edit_menu)
        edit_menu.add_command(label="Cut    ", command=lambda: cut_text(False), accelerator="(Ctrl+x)")
        edit_menu.add_command(label="Copy   ", command=lambda: copy_text(False), accelerator="(Ctrl+c)")
        edit_menu.add_command(label="Paste  ", command=lambda: paste_text(False), accelerator="(Ctrl+v)")
        edit_menu.add_separator()
        edit_menu.add_command(label="Undo", command=text.edit_undo, accelerator="(Ctrl+z)")
        edit_menu.add_command(label="Redo", command=text.edit_redo, accelerator="(Ctrl+y)")

        #color menu

        color_menu = Menu(my_menu, tearoff=False)
        my_menu.add_cascade(label="Colors", menu=color_menu)
        color_menu.add_command(label="Selected Text", command=text_color)
        color_menu.add_command(label="All Text", command=all_text_color)
        color_menu.add_command(label="Background", command=bg_color)


        #menu chevave

        options_menu = Menu(my_menu, tearoff=False)
        my_menu.add_cascade(label="Options", menu=options_menu)
        options_menu.add_command(label="Nigth Mode On", command=night_on)
        options_menu.add_command(label="Nigth Mode Off", command=night_off)


        #bottom
        status_bar = Label(self, text='Ready        ')
        status_bar.pack(fill=X, side=BOTTOM, ipady=15)

        self.bind('<Control-Key-x>', cut_text)
        self.bind('<Control-Key-c>', copy_text)
        self.bind('<Control-Key-v>', paste_text)




        #button bold
        bold_button = Button(toolbar_frame, text="Bold", command=bold, bg='#ADD8E6')
        bold_button.grid(row=0, column=0, sticky=W, padx=5)
        #button italik
        italic_button = Button(toolbar_frame, text="Italic", command=italic, bg='#ADD8E6')
        italic_button.grid(row=0, column=1, padx=5)
        #undo/redo
        undo_button = Button(toolbar_frame, text="Undo", command=text.edit_undo, bg='#ADD8E6')
        undo_button.grid(row=0, column=2, padx=5)
        redo_button = Button(toolbar_frame, text="Redo", command=text.edit_redo, bg='#ADD8E6')
        redo_button.grid(row=0, column=3, padx=5)

        #coleur text
        color_text_button = Button(toolbar_frame, text="Text Color", command=text_color, bg='#ADD8E6')
        color_text_button.grid(row=0, column=4, padx=5)
        #button titre
        titre_button = Button(toolbar_frame, text="Titre", command=titre, bg='#ADD8E6')
        titre_button.grid(row=0, column=5, padx=5)
        #sous titre
        sous_titre_button = Button(toolbar_frame, text="Sous Titre", command=sous_titre, bg='#ADD8E6')
        sous_titre_button.grid(row=0, column=6, padx=5)
        #normal
        normal_button = Button(toolbar_frame, text="Normal", command=normal, bg='#ADD8E6')
        normal_button.grid(row=0, column=7, padx=5)
        #sousligne
        sousligne_button = Button(toolbar_frame, text="Souligne", command=sousligne, bg='#ADD8E6')
        sousligne_button.grid(row=0, column=8, padx=5)
        #strike
        strike_button = Button(toolbar_frame, text="Strike", command=strike, bg='#ADD8E6')
        strike_button.grid(row=0, column=9, padx=5)
        ###bold_italic
        bold_italic_button = Button(toolbar_frame, text="Bold_Italic", command=bold_italic, bg='#ADD8E6')
        bold_italic_button.grid(row=0, column=10, padx=5)
        ######bold_sousligne
        b_s = Button(toolbar_frame, text="Bold_Souligne", command=b_s, bg='#ADD8E6')
        b_s.grid(row=0, column=11, padx=5)
        ############bold_strik
        b_st =Button(toolbar_frame, text="Bold_Strike", command=b_st, bg='#ADD8E6')
        b_st.grid(row=0, column=12, padx=5)
        ###########italic_souling
        it_s =Button(toolbar_frame, text="Italic_Souligne", command=it_s, bg='#ADD8E6')
        it_s.grid(row=0, column=13, padx=5)
        ##italic_strike
        it_sk =Button(toolbar_frame, text="Italic_Strike", command=it_sk, bg='#ADD8E6')
        it_sk.grid(row=0, column=14, padx=5)
        ##italic_strike
        b_it_s =Button(toolbar_frame, text="Bold_Italic_souligne", command=b_it_s, bg='#ADD8E6')
        b_it_s.grid(row=0, column=15, padx=5)
        ##italic_str
        b_it_sk =Button(toolbar_frame, text="Bold_Italic_strike", command=b_it_sk, bg='#ADD8E6')
        b_it_sk.grid(row=0, column=16, padx=5)
        ##ital
        b_it_sk =Button(toolbar_frame, text="Bold_Italic_strike", command=b_it_sk, bg='#ADD8E6')
        b_it_sk.grid(row=0, column=16, padx=5)
        ##ital
        size_f =Button(toolbar_frame, text="Size", command=size, bg='#ADD8E6')
        size_f.grid(row=0, column=17, padx=5)
    
#################33





r = Tk()
r.geometry('1350x700')
r.title('login page')
r.configure(bg='#2C4762')
r.resizable(False,False)
r.state('zoomed')
#r.wm_attributes('-transparentcolor', '#DD4B39')
c = Frame(r, bg='cadetblue'  , width='950', height='600')
#log.attributes('-alpha' ,0.5)
c.place(x=200, y=70)

s=ttk.Style()
s.theme_use('clam')
s.configure("red.Horizontal.TProgressbar",foreground='red')
progress=Progressbar(c,style="red.Horizontal.TProgressbar",orient=HORIZONTAL,length=950,mode='determinate')

###################3
def bar():

    l4=Label(c,text='Loading...',fg='white', bg='cadetblue')
    lst4=('Calibri (Body)', 10)
    l4.configure(font=lst4)
    l4.place(x=0,y=550)

    
    import time
    r=0
    for i in range(100):
        progress['value']=r
        c.update_idletasks()
        time.sleep(0.01)
        r=r+1
    c.destroy()
    MainApp()
progress.place(x=-10,y=580)

Frame(c,width=750,height=450,bg='cadetblue').place(x=0,y=0)
b1=Button(c,width=10,height=1,text='Get Started', command=bar,border=0,fg='#000000',bg='#778899')
b1.place(x=400,y=500)
####################
c1 = Label(c, text='WELCOME TO MY APP', font=('yu gothic ui', 35, 'bold'), bg='cadetblue', fg='#000000')
c1.place(x=250, y=50)


f = Font(size=50, weight='bold', slant='italic')
c2 = Label(c, text='MARKDOWN', font=f, bg='cadetblue',fg='#000000')
c2.place(x=250, y=300)

r.mainloop()
```
