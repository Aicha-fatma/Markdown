```python
from tkinter import *
from tkinter.font import Font
from tkinter import font , filedialog
from tkinter import messagebox as mbox
from markdown2 import Markdown
from tkhtmlview import HTMLLabel




root = Tk()
root.geometry("600x720")
###################


###############
fonte_texte = Font(family="Times", size=12)
fonte_italique = Font(family="Times", slant="italic", size=12)
fonte_gras = Font(family="Times", weight="bold", size=12)
fonte_titre = Font(family="Times", weight="bold", size=20)
fonte_sous_titre = Font(family="Times", weight="bold", size=16)



class App():
     def afficher_ligne_v2(par,posy):
        mode = "texte"
        indentation = 40
        if par[0:2] == "##": 
            mode = "sous_titre"
            par = par[2:] 
        elif par[0] == "#": 
            mode = "titre"
            par = par[1:] 
        elif par[0] == "+": 
            mode = "liste"
            par = u'\u2022' + par[1:] 
            indentation = 40
     
        posx = indentation
        liste_mots = par.split()
        for mot in liste_mots:
            mot = mot + " " 
            mot_canvas = canvas.create_text(posx,posy, text=mot,anchor=NW,font=fonte_titre,fill="lightslategrey")
            posx = canvas.bbox(mot_canvas)[2]
    
     def choix_fonte(mode,en_gras,en_italique):
            if mode == "titre":
                fonte = fonte_titre
            elif mode == "sous_titre":
                fonte = fonte_sous_titre
            else: # Mode texte ou liste
                if en_gras:
                    fonte = fonte_gras
                elif en_italique:
                     fonte = fonte_italique
                else:
                    fonte = fonte_texte
            return fonte
    
     def afficher_ligne_v3(par,posy):
        mode = "texte"
        indentation = 20
        if par[0:2] == "##":
            mode = "sous_titre"
            par = par[2:] 
        elif par[0] == "#":
            mode = "titre"
            par = par[1:] 
        elif par[0] == "+":
            mode = "liste"
            par = u'\u2022' + par[1:] 
            indentation = 40

        en_gras = False
        en_italique = False

        posx = indentation
        liste_mots = par.split()
        for mot in liste_mots:
            if mot == "**": 
                en_gras = not(en_gras)
                mot = ""
            if en_gras:
                fonte = fonte_gras
            if mot == "*": 
                en_italique = not(en_italique)
                mot = ""
            fonte = choix_fonte(mode,en_gras,en_italique)
            if mot != "":
                mot = mot + " "
            mot_canvas = canvas.create_text(posx,posy, text=mot,anchor=NW,font=fonte_titre,fill="blue")
            posx = canvas.bbox(mot_canvas)[2]
        return
     

#################

my_frame = Frame(root)
my_frame.pack(pady=5)

text_scroll = Scrollbar(my_frame)
text_scroll.pack(side=RIGHT, fill=Y)


my_text = Text(my_frame, width=97, height=25, font=("Helvetica", 16),background="linen", undo=True,
               yscrollcommand=text_scroll.set)
my_text.pack()
               
 
text_scroll.config(command=my_text.yview)
def log():
    main = App()

btn = Button(text="Entre",bg="#F3A712",font=("Courirer", 18), command=log ).place(x=30, y=380,width=350, height=50)






#################
def main():
    app = App()
    app.mainloop()
               
            
    if "__main__" == __name__:
        main()
root.mainloop()
```
